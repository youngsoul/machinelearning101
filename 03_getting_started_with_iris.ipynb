## Loading the iris dataset into scikit-learn

```python
# matplotlib setup
# Print figures in the notebook
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.colors import ListedColormap
# Import patch for drawing rectangles in the legend
from matplotlib.patches import Rectangle

# Create color maps
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# Create a legend for the colors, using rectangles for the corresponding colormap colors
labelList = []
for color in cmap_bold.colors:
    labelList.append(Rectangle((0, 0), 1, 1, fc=color))
```

```python
# import load_iris function from datasets module
from sklearn.datasets import load_iris
```

```python
# load iris dataset and its attributes
iris = load_iris()
```

```python
# look at all of the possible keys from the iris dataset Bunch
iris.keys()
```

```python
print(f"Feature Names: {iris.feature_names}")
print(f"Data Shape: {iris.data.shape}")
print(f"Target names: {iris.target_names}")
print(f"Target numbers: {set(iris.target)}")
print(f"Target Shape: {iris.target.shape}")
```

```python
print(iris.DESCR)
```

```python
# print the iris data.  These are considered Features
print(f"Feature Names: {iris.feature_names}")
print(iris.data)
```

### Feature Matrix by convention is called:  X

### Response or label by convention is called:  y

```python
# store feature matrix in "X"
# X is 150 sample rows by 4 feature columns
X = iris.data

# store response vector in "y"
# Y is 150 target values
y = iris.target
```

# Visualizing the data

Visualizing the data can help us better understand the data and make use of it. The following block of code will create a plot of sepal length (x-axis) vs sepal width (y-axis). The colors of the datapoints correspond to the labeled species of iris for that example.

After plotting, look at the data. What do you notice about the way it is arranged?


```python
# Plot the data for Sepal length vs Sepal Width

# Sepal length and width
X_sepal = X[:,:2]  # X[rows, columns]  X[:, :2] all of the rows (:), and columns 0-2 (exclusive of 2)

# Get the minimum and maximum values with an additional 0.5 border
x_min, x_max = X_sepal[:, 0].min() - .5, X_sepal[:, 0].max() + .5
y_min, y_max = X_sepal[:, 1].min() - .5, X_sepal[:, 1].max() + .5

plt.figure(figsize=(8, 6))

# Plot the training points
plt.scatter(X_sepal[:, 0], X_sepal[:, 1], c=y, cmap=cmap_bold)
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')
plt.title('Sepal width vs length')

# Set the plot limits
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.legend(labelList, iris.target_names)

plt.show()

```

```python
# Plot the data for petal length and petal width

# Sepal length and width
X_petal = X[:,2:4] # all rows, columns 2-4 exclusive of 4

# Get the minimum and maximum values with an additional 0.5 border
x_min, x_max = X_petal[:, 0].min() - .5, X_petal[:, 0].max() + .5
y_min, y_max = X_petal[:, 1].min() - .5, X_petal[:, 1].max() + .5

plt.figure(figsize=(8, 6))

# Plot the training points
plt.scatter(X_petal[:, 0], X_petal[:, 1], c=y, cmap=cmap_bold)
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.title('Petal width vs length')

# Set the plot limits
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.legend(labelList, iris.target_names)

plt.show()


```

## Machine learning terminology

- Each row is an **observation** (also known as: sample, example, instance, record)
- Each column is a **feature** (also known as: predictor, attribute, independent variable, input, regressor, covariate)

```python
# print integers representing the species of each observation
print("setosa = 0, versicolor=1, virginica=2")
print(iris.target)
```

```python
# Use Pandas to read the data into a Pandas Dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = pd.Series(iris.target)
# Display every 30th value
df[::30]
```

- Each value we are predicting is the **response** (also known as: target, outcome, label, dependent variable)
- **Classification** is supervised learning in which the response is categorical
- **Regression** is supervised learning in which the response is ordered and continuous

```python
# Print out information on the DataFrame
df.info()
```

```python
# Print out information about the data in the DataFrame
df.describe()
```

```python
print("setosa = 0, versicolor=1, virginica=2")
plot = sns.pairplot(df, hue='target')
```

```python
sns.pairplot(df, kind='reg')
```

## Requirements for working with data in scikit-learn

1. Features and response are **separate objects**.  X represents Features, y represents response
2. Features and response should be **numeric**
3. Features and response should be **NumPy arrays**
4. Features and response should have **specific shapes**

```python
# check the types of the features and response
print(type(iris.data))
print(type(iris.target))
```

```python
# check the shape of the features (first dimension = number of observations, second dimensions = number of features)
print(iris.data.shape)
```

```python
# check the shape of the response (single dimension matching the number of observations)
print(iris.target.shape)
```
